```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
##calculate clustering by genotype
```

```python
dir_ = '/Volumes/My_Passport/cnmfe_analysis_files/OpenFieldAnalysis/'
ko_mice = ['GRIN011', 'GRIN018', 'GRIN012']
wt_mice = ['GRIN013', 'GRIN026', 'GRIN027']
```

```python
ko_distances = []
ko_pearsons = []
for mouse in ko_mice:
    ko_distances.append(pd.read_hdf(str(dir_  + mouse  + '_euclidean_distance.h5')))
    ko_pearsons.append(pd.read_hdf(str(dir_  + mouse   + '_pairwise_pearson.h5'))) 

ko_distances_df = pd.concat(ko_distances, axis=1, keys=ko_mice)
ko_pearsons_df = pd.concat(ko_pearsons, axis=1, keys=ko_mice)
```

```python
wt_distances = []
wt_pearsons = []
for mouse in wt_mice:
    wt_distances.append(pd.read_hdf(str(dir_  + mouse  + '_euclidean_distance.h5')))
    wt_pearsons.append(pd.read_hdf(str(dir_  + mouse   + '_pairwise_pearson.h5'))) 

wt_distances_df = pd.concat(wt_distances, axis=1, keys=wt_mice)
wt_pearsons_df = pd.concat(wt_pearsons, axis=1, keys=wt_mice)
```

```python
plt.scatter(ko_distances_df, ko_pearsons_df)
plt.show()
```

```python
plt.scatter(wt_distances_df, wt_pearsons_df)
plt.show()
```
