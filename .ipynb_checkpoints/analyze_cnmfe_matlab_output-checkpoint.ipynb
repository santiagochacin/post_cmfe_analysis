```python
mat_results = sio.loadmat(full_path_to_file)
```

```python
mat_results['A']
```

```python
results = mat_results
# Call contour plotting function from Caiman (with our results from MATLAB!)
coordinates = utils.plot_contours(results['A'].todense(), 
                                      results['Cn'],
                                      display_numbers=False, maxthr=.6,
                                      cmap='gray', colors='r')
#plt.title(name)
plt.axis('off')
```

```python
len(coordinates)
```

```python
#create list of tuples for neuron pairs
neuron_pairs = []
for i in range(len(coordinates)):
    for j in range(len(coordinates)):
        if i != j: 
            neuron_pairs.append((i,j))
            
```

```python
pairwise = {}
for pair in tqdm(neuron_pairs):
    #calculate pearsons r
    r = stats.pearsonr(mat_results['C'][pair[0]], mat_results['C'][pair[1]])
    #calc euclidian distance 
    distance = dist.euclidean(coordinates[pair[0]]['CoM'], coordinates[pair[1]]['CoM'])
    pairwise[pair] = (r, distance)
```

```python
r_coeffs = [item[0][0] for item in pairwise.values()]
euclidian_distances = [item[0][1] for item in pairwise.values()]
```

```python
plt.plot(euclidian_distances, r_coeffs)
```

```python
z = np.polyfit(euclidian_distances, r_coeffs, 2)
```

```python
plt.plot(z)
```
